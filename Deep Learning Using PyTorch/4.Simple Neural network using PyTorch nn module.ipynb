# Simple neural network using PyTorch nn module (single neuron)

# Improvements

1. Building the neural network using nn module

2. Using built-in activation function

3. Using built-in loss function

4. Using built-in optimizer

```python
# create model class
import torch
import torch.nn as nn


class Model(nn.Module): #extend nn module to use inside my model

  def __init__(self, num_features):
    super().__init__() # nn.module class invoked
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()


  def forward(self, features):
    out = self.linear(features) # z = wx+b
    out = self.sigmoid(out) # prediction sigmoid(z)
    return out
```

```python
# create dataset
features = torch.rand(10, 5)

# create model
model = Model(features.shape[1])

# call model for forward pass
model(features) # similar to model.forward(features)
```

# Complex neural network. 5 neurons input layer, 3 neurons hidden layer and one neuron in output layer

hidden layer relu and output layer sigmoid

```python
class Model(nn.Module):

  def __init__(self, num_features):

    super().__init__()

    self.linear1 = nn.Linear(num_features, 3) # input: features, output: 3 neurons
    self.relu = nn.ReLU()

    self.linear2 = nn.Linear(3, 1)
    self.sigmoid = nn.Sigmoid()


  def forward(self, features):
    out = self.linear1(features)
    out = self.relu(out)

    out = self.linear2(out)
    out = self.sigmoid(out)

    return out
```

```python
# create dataset
features = torch.rand(10, 5)

# create model
model = Model(features.shape[1])

# call model for forward pass
model(features)
```

# Sequential neural network

```python
class Model(nn.Module):

  def __init__(self, num_features):

    super().__init__()
    self.network = nn.Sequential (
      nn.Linear(num_features, 3),
      nn.ReLU(),
      nn.Linear(3, 1),
      nn.Sigmoid()
      )



  def forward(self, features):
    out = self.network(features)
    return out
```

```python
# create dataset
features = torch.rand(10, 5)

# create model
model = Model(features.shape[1])

# call model for forward pass
model(features)
```

# Training pipeline with real data using nn module

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn


df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.shape
```

```python
# import necessary libary functions
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder


print(df.shape)

df.head()
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)

print(df.shape)
df.head()
```

# Train test split and preprocessing

```python
# train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.3)

# preprocessing
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)


print(X_train)

y_train
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.fit_transform(y_test)

print(y_train)

print(y_test)
```

```python
X_train_tensor = torch.from_numpy(X_train.astype(np.float32))
X_test_tensor = torch.from_numpy(X_test.astype(np.float32))
y_train_tensor = torch.from_numpy(y_train.astype(np.float32))
y_test_tensor = torch.from_numpy(y_test.astype(np.float32))
```

```python
X_train_tensor.shape
```

```python
y_train_tensor.shape
```

# Define model

```python
import torch.nn as nn
class MySimpleNN(nn.Module):

  def __init__(self, num_features):
    super().__init__()

    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()


  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)

    return out
```

# Importants parameters

```python
learning_rate = 0.1
epochs = 25
```

```python
loss_function = nn.BCELoss()
type(loss_function)
```

# Training the model

```python
# create model
model = MySimpleNN(X_train_tensor.shape[1])


# define optimzer
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

# define loop
for epoch in range(epochs):

  # forward pass
  y_pred = model(X_train_tensor)

  # loss calculation
  loss = loss_function(y_pred, y_train_tensor.view(-1, 1))

  # gradient clear
  optimizer.zero_grad()

  # backward pass
  loss.backward()

  # parameters update
  optimizer.step()

  # print loss in each epoch
  print(f"epoch: {epoch + 1}, Loss: {loss.item()}")
```

# Model evaluation

```python
# model evaluation
with torch.no_grad():
  y_pred = model.forward(X_train_tensor)
  y_pred = (y_pred > 0.5).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f"Accuracy: {accuracy.item()}")
```
