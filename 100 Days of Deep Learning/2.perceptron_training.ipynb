# How perceptron update bias and weight

✅ What Happens for Each Row?
For each data of student

The perceptron predicts placement using current weights and bias.

Compares prediction with actual result.

If incorrect, updates weights and bias to reduce future error.

Moves to next student.

This process is typically repeated for many epochs (passes over the entire dataset).

Note: to see graph visualization positive and negative area changes, go to
https://www.desmos.com/calculator with 2x+3y+3=0

** After 1 epoch:**

You updated weights and bias once based on incorrect predictions.

But what if now the first row is misclassified with new weights?

So you repeat: go back to row 1 → forward pass → maybe update → then row 2 → and so on.

# Training Perceptron

```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=10)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', s=100)
```

### Perceptron Training Function

```python
def step(x):
    return 1 if x >=0 else 0
```

```python
def perceptron(X, y):
    X = np.insert(X, 0, 1, axis=1)   
    weights = np.ones(X.shape[1])   
    lr = 0.1
    
    for i in range(1000):
        j = np.random.randint(0, 100)   
        y_pred = step(np.dot(X[j], weights))
        weights = weights + lr * (y[j] - y_pred) * X[j]
    
    return weights[0], weights[1:]
```

```python
intercept_, coef_ = perceptron(X, y)

print(coef_, intercept_)
```

```python
m = -(coef_[0] / coef_[1])
b = -(intercept_ / coef_[1])
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m * x_input + b
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_input, y_input, color='red', label='Decision Boundary')
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
plt.ylim(-3, 2)
```
